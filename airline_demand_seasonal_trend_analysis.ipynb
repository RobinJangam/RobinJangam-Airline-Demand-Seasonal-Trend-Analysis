```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

# Set a style for the plots for better aesthetics
sns.set_style("whitegrid")

# --- Expanded Basic Project with Diverse Charts (excluding heatmaps, Age by Month, and Age by Flight Status Box Plots, and Gender Distribution): Analysis of Airline Passenger Trends and Characteristics in 2022 ---

print("--- Starting Expanded Basic Project with Diverse Charts (excluding heatmaps, Age by Month, and Age by Flight Status Box Plots, and Gender Distribution): Analysis of Airline Passenger Trends and Characteristics in 2022 ---")

# Data Loading and Cleaning (assuming the file 'Airline Dataset.csv' is available)
try:
    df1 = pd.read_csv('Airline Dataset.csv', on_bad_lines='skip')
    print("Successfully loaded 'Airline Dataset.csv'")

    # Cleaning steps from previous analysis
    df1['Last Name'] = df1['Last Name'].fillna(df1['Last Name'].mode()[0])
    df1['Gender'] = df1['Gender'].fillna(df1['Gender'].mode()[0])
    df1['Age'] = pd.to_numeric(df1['Age'], errors='coerce')
    df1['Age'] = df1['Age'].fillna(df1['Age'].median())
    df1['Flight Status'] = df1['Flight Status'].fillna(df1['Flight Status'].mode()[0])
    df1['Departure Date'] = pd.to_datetime(df1['Departure Date'], errors='coerce')
    df1['Departure Month'] = df1['Departure Date'].dt.month
    df1['Departure Day of Week'] = df1['Departure Date'].dt.dayofweek # Add Day of Week

    print("Data cleaning and preparation complete.")
    print("Missing values after cleaning:\n", df1.isnull().sum())
    print("Data types after cleaning:\n", df1.dtypes)

    # --- Display Key Data Summaries ---
    print("\n--- Key Data Summaries ---")
    print("Shape of the DataFrame:", df1.shape)
    print("\nData Types:\n", df1.dtypes)
    print("\nDescriptive Statistics for Numerical Columns:\n", df1.describe())
    print("\nMissing Values per Column:\n", df1.isnull().sum())
    print("\nUnique Values in Categorical Columns:")
    for col in ['Gender', 'Flight Status']:
        print(f"- {col}: {df1[col].unique()}")


    # --- Analyze Monthly Passenger Trends ---
    print("\n--- Analyzing Monthly Passenger Trends for 2022 ---")

    # Calculate monthly passenger counts
    monthly_passengers = df1.groupby('Departure Month')['Passenger ID'].count().reset_index()
    monthly_passengers.rename(columns={'Passenger ID': 'Passenger Count'}, inplace=True)
    print("Monthly Passenger Counts:")
    display(monthly_passengers)

    # Visualize monthly passenger counts (Line plot remains useful for trends)
    plt.figure(figsize=(12, 7))
    sns.lineplot(x='Departure Month', y='Passenger Count', data=monthly_passengers, marker='o', linewidth=2.5)
    plt.title('Total Passengers per Month in 2022', fontsize=16)
    plt.xlabel('Departure Month', fontsize=12)
    plt.ylabel('Number of Passengers', fontsize=12)
    plt.xticks(range(1, 13), fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

    # --- Analyze Trends by Day of Week ---
    print("\n--- Analyzing Passenger Trends by Day of Week in 2022 ---")

    # Calculate passenger counts by day of week
    weekday_passengers = df1.groupby('Departure Day of Week')['Passenger ID'].count().reset_index()
    weekday_passengers.rename(columns={'Passenger ID': 'Passenger Count'}, inplace=True)
    print("Passenger Counts by Day of Week:")
    display(weekday_passengers)

    # Visualize passenger counts by day of week (Bar plot is good for comparison)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Departure Day of Week', y='Passenger Count', data=weekday_passengers, palette='viridis')
    plt.title('Total Passengers by Day of Week in 2022', fontsize=16)
    plt.xlabel('Day of Week (0=Monday, 6=Sunday)', fontsize=12)
    plt.ylabel('Number of Passengers', fontsize=12)
    plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

    # --- Analyze Flight Status Distribution by Month (Stacked Bar Chart) ---
    print("\n--- Analyzing Flight Status Distribution by Month in 2022 ---")

    # Prepare data for stacked bar chart
    monthly_flight_status = df1.groupby(['Departure Month', 'Flight Status']).size().unstack(fill_value=0)
    display(monthly_flight_status)

    # Visualize flight status distribution by month (Stacked Bar Chart)
    plt.figure(figsize=(14, 8))
    monthly_flight_status.plot(kind='bar', stacked=True, colormap='plasma', ax=plt.gca())
    plt.title('Flight Status Distribution by Month in 2022', fontsize=16)
    plt.xlabel('Departure Month', fontsize=12)
    plt.ylabel('Number of Flights', fontsize=12)
    plt.xticks(rotation=0, fontsize=10)
    plt.yticks(fontsize=10)
    plt.legend(title='Flight Status', fontsize=10)
    plt.tight_layout()
    plt.show()

    # --- Analyze Distribution of Top Airports ---
    print("\n--- Analyzing Distribution of Top Airports in 2022 ---")

    plt.figure(figsize=(12, 8))
    top_n_airports = df1['Airport Name'].value_counts().nlargest(20).index # Top 20 airports
    sns.countplot(y='Airport Name', data=df1[df1['Airport Name'].isin(top_n_airports)], order=top_n_airports, palette='viridis')
    plt.title('Top 20 Airports by Number of Flights in 2022', fontsize=16)
    plt.xlabel('Number of Flights', fontsize=12)
    plt.ylabel('Airport Name', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

    # --- Analyze Distribution of Top Countries ---
    print("\n--- Analyzing Distribution of Top Countries in 2022 ---")

    plt.figure(figsize=(10, 7))
    top_n_countries = df1['Country Name'].value_counts().nlargest(15).index # Top 15 countries
    sns.countplot(y='Country Name', data=df1[df1['Country Name'].isin(top_n_countries)], order=top_n_countries, palette='plasma')
    plt.title('Top 15 Countries by Number of Flights in 2022', fontsize=16)
    plt.xlabel('Number of Flights', fontsize=12)
    plt.ylabel('Country Name', fontsize=12)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()

    # --- Identify Peak/Off-Peak Months (from Monthly Trends Analysis) ---
    print("\n--- Identifying Peak/Off-Peak Months ---")

    if not monthly_passengers.empty:
        peak_month = monthly_passengers.loc[monthly_passengers['Passenger Count'].idxmax()]
        off_peak_month = monthly_passengers.loc[monthly_passengers['Passenger Count'].idxmin()]

        print(f"Peak Month (highest passengers): Month {int(peak_month['Departure Month'])} ({int(peak_month['Passenger Count'])} passengers)")
        print(f"Off-Peak Month (lowest passengers): Month {int(off_peak_month['Departure Month'])} ({int(off_peak_month['Passenger Count'])} passengers)")
    else:
        print("Could not identify peak/off-peak months as monthly passenger data is empty.")

    # --- Summarize Findings ---
    print("\n--- Summary of Findings from Expanded Basic Project ---")
    print("Based on the 2022 dataset:")
    if not monthly_passengers.empty:
        print(f"- Monthly passenger trends show that Month {int(peak_month['Departure Month'])} had the highest number of passengers, and Month {int(off_peak_month['Departure Month'])} had the lowest.")
        print("- The line plot illustrates the variation in passenger numbers across the months.")
    if not weekday_passengers.empty:
        print("- Passenger counts also vary by day of the week, indicating weekly patterns.")
    print("- The stacked bar chart shows the breakdown of flight statuses within each month.")
    print("- The bar charts highlight the top airports and countries by flight volume.")
    print("Note: This analysis is based on a single year (2022) and may not reflect long-term seasonal trends or year-over-year changes.")


except FileNotFoundError:
    print("Error: 'Airline Dataset.csv' not found. Please make sure the file is uploaded to your Colab environment.")
except Exception as e:
    print(f"An error occurred during the project execution: {e}")

print("\n--- Expanded Basic Project with Diverse Charts Execution Complete ---")
```
